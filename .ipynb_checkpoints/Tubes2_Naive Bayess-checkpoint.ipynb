# Inisialisasi Library

```python
#Datasets Library
import pandas as pd
from pandas import DataFrame
import numpy as np
import graphviz
from sklearn.externals import joblib

# Algorithm Library
from sklearn.naive_bayes import GaussianNB
from sklearn import tree
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier

#library for computing performance and confusion matrix
from sklearn.metrics import confusion_matrix 
from sklearn.metrics import accuracy_score 
from sklearn.metrics import average_precision_score 
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import confusion_matrix

#library for trainig types
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_predict

#library for preprocessing data
from collections import defaultdict
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import Imputer

import random
```

## Load Dataset

```python
atributeNames = ["age","workclass","fnlwgt","education","education-num","marital-status","ocupation","relationship","race","sex","capital-gain","capital-loss","hours-per-week","native-country","target"]
income = pd.read_csv("CencusIncome.data.txt",names = atributeNames);
incomeTest = pd.read_csv("CencusIncome.test.txt",names = atributeNames);

print(income)
print(incomeTest)
```

# Preprocessing Data

```python
#pd.get_dummies
columnCount = None
incomeDummies = None
for columnCount in income.columns.values:
    if ((columnCount!="age")and(columnCount!="fnlwgt")and(columnCount!="education-num")and(columnCount!="capital-gain")and(columnCount!="capital-loss")and(columnCount!="hours-per-week")):
        tempData = pd.get_dummies(income[columnCount])
        tempData = tempData.set_index(income.index.values)
        incomeDummies = pd.concat([incomeDummies,tempData],axis=1)
        tempData = None
    else :
        incomeDummies = pd.concat([incomeDummies,income[columnCount]],axis=1)

columnCount = None
incomeDummiesTest = None
for columnCount in incomeTest.columns.values:
    if ((columnCount!="age")and(columnCount!="fnlwgt")and(columnCount!="education-num")and(columnCount!="capital-gain")and(columnCount!="capital-loss")and(columnCount!="hours-per-week")):
        tempData = pd.get_dummies(incomeTest[columnCount])
        tempData = tempData.set_index(incomeTest.index.values)
        incomeDummiesTest = pd.concat([incomeDummiesTest,tempData],axis=1)
        tempData = None
    else :
        incomeDummiesTest = pd.concat([incomeDummiesTest,incomeTest[columnCount]],axis=1)

print(incomeDummies)
print(incomeDummiesTest)
incomeDummiesData = incomeDummies.iloc[:,0:108].values
incomeDummiesTarget = incomeDummies.iloc[:,108:110].values
incomeDummiesTestData = incomeDummiesTest.iloc[:,0:107].values
incomeDummiesTestTarget = incomeDummiesTest.iloc[:,107:109].values
print(incomeDummiesData)
print(incomeDummiesTarget)
print(incomeDummiesTestData)
print(incomeDummiesTestTarget)
```

```python
#LabelEncoder
categoryEncode = preprocessing.LabelEncoder()
columnCount = None
incomeLabel = pd.read_csv("CencusIncome.data.txt",names = atributeNames);
incomeLabelTest = pd.read_csv("CencusIncome.test.txt",names = atributeNames);
for columnCount in income.columns.values:
    if income[columnCount].dtypes=='object':
        tempData = income[columnCount]
        categoryEncode.fit(tempData.values)
        incomeLabel[columnCount] = categoryEncode.transform(income[columnCount])
        tempData = None
    else :
        incomeLabel[columnCount] = income[columnCount]
        
columnCount = None
for columnCount in incomeTest.columns.values:
    if incomeTest[columnCount].dtypes=='object':
        tempData2 = incomeTest[columnCount]
        categoryEncode.fit(tempData2.values)
        incomeLabelTest[columnCount] = categoryEncode.transform(incomeTest[columnCount])
        tempData2 = None
    else :
        incomeLabelTest[columnCount] = incomeTest[columnCount]

print(incomeLabel)
print(incomeLabelTest)
incomeLabelData = incomeLabel.iloc[:,0:14].values
incomeLabelTarget = incomeLabel.iloc[:,14].values
incomeLabelTestData = incomeLabelTest.iloc[:,0:14].values
incomeLabelTestTarget = incomeLabelTest.iloc[:,14].values
print(incomeLabelData)
print(incomeLabelTarget)
print(incomeLabelTestData)
print(incomeLabelTestTarget)
```

## 10-Fold Cross Validation

```python
np.random.seed(0)
indices = np.random.permutation(32561)

foldLabelTrainData = incomeLabelData[indices[:]]
foldLabelTrainTarget = incomeLabelTarget[indices[:]]

print(foldLabelTrainData)
print(foldLabelTrainTarget)
```

### kNN

### Naive Bayes

```python
foldNaiveBayes = GaussianNB()
foldNaiveBayes.fit(foldLabelTrainData,foldLabelTrainTarget)
```

```python
#Script for 10-fold training and measuring performance

#This is the general score from 10 iteration
#cv is the k of k-fold training
NaiveBayes_scores = cross_val_score(foldNaiveBayes, foldLabelTrainData, foldLabelTrainTarget, cv = 10)

#This is the accuracy score of each iterationa
NaiveBayes_scores_acc = cross_val_score(foldNaiveBayes, foldLabelTrainData, foldLabelTrainTarget, cv = 10, scoring = 'accuracy')
print("Accuracy: %0.2f (+/- %0.2f)" % (NaiveBayes_scores_acc.mean(), NaiveBayes_scores_acc.std() * 2))

#The precision score
NaiveBayes_scores_prec = cross_val_score(foldNaiveBayes, foldLabelTrainData, foldLabelTrainTarget, cv = 10, scoring = 'precision_micro')
print("Precision: %0.2f (+/- %0.2f)" % (NaiveBayes_scores_prec.mean(), NaiveBayes_scores_prec.std() * 2))

#The recall score
NaiveBayes_scores_recall = cross_val_score(foldNaiveBayes, foldLabelTrainData, foldLabelTrainTarget, cv = 10, scoring = 'recall_micro')
print("Precision: %0.2f (+/- %0.2f)" % (NaiveBayes_scores_recall.mean(), NaiveBayes_scores_recall.std() * 2))
```

```python
foldNaiveBayesPredict = foldNaiveBayes.predict(incomeLabelTestData)
#generate confusion_matrix
print((pd.crosstab(incomeLabelTestTarget, foldNaiveBayesPredict, rownames=['True'], colnames=['Predicted'], margins=True)))
```
